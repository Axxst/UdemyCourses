```python
JB_Movies = pd.read_csv('jamesbond.csv', index_col='Film')
JB_Movies.head()
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == 'Pierce Brosnan':
        return 'The best'
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    else:
        return 'I have no clue'

JB_Movies.apply(good_movie, axis= 'columns')
```
