```python
import pandas as pd
```

```python
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors = fortune.groupby(by='Sector')
fortune.head()
```

```python
sectors.agg({'Revenue': 'sum',
             'Profits': 'sum',
            'Employees': 'mean'})
```

```python
for data, sector in sectors:
    Highest_revenue = sector.nlargest(1, 'Revenue')

df =  pd.DataFrame(columns= fortune.columns)
df= df.append(Highest_revenue)
```

```python
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors = fortune.groupby(by='Sector')
fortune.head()
```
